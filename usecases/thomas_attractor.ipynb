```python
import os
os.chdir('..')

from functools import partial

from matplotlib.animation import FuncAnimation
from matplotlib.pyplot import show as plt_show

from attractor import StrangeAttractor, initialize_animation, update_location

import warnings
warnings.filterwarnings('ignore')
```

```python
%matplotlib notebook
```

```python
strange_attractor = StrangeAttractor(
    attractor_type='thomas', 
    method='runge-kutta', 
    num_iterations=100000,
    step=0.01,
)
strange_attractor.set_init_coordinates(init_coordinates=[1.1, 1.1, -0.01])
strange_attractor.build_attractor(params=[0.208186])
```

```python
strange_attractor.plot_projection(projection='xoy')
```

```python
strange_attractor.plot_projection(projection='xoz')
```

```python
strange_attractor.plot_projection(projection='yoz')
```

```python
fig, ax, scatter, line = initialize_animation(data=strange_attractor.cached_movement)

keep_iterations = [x for x in range(len(strange_attractor.cached_movement)) if x%10 == 0]
data = strange_attractor.cached_movement[keep_iterations, :]

ani = FuncAnimation(
    fig, 
    partial(
        update_location, 
        data=data, 
        ax=ax, 
        line=line, 
        scatter=scatter,
    ), 
    frames=len(data), 
    interval=2,
)
plt_show()
```
